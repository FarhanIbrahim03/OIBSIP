```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
#importing dataset

dat=pd.read_csv('C:/Users/Farhan/Documents/python jupyter/Datasets/Advertising.csv')
dat.head()
```

```python
#seems like unnamed column doesnt provide any significance so we will drop it
dat_new=dat.drop('Unnamed: 0',axis=1)
dat_new
```

```python
#splitting data

x=dat_new.iloc[:,:-1]
y=dat_new.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
#ml model

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=50)
```

```python
lr=LinearRegression()

lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
y_pred
```

```python
#metricsabs
from sklearn.metrics import r2_score

r2_score(y_pred,y_test)
```

```python
figsize=(30, 30)

ax1 = sns.distplot(dat_new['Sales'], hist=False, color="r", label="Actual Value")
sns.distplot(y_pred, hist=False, color="g", label="Fitted Values" , ax=ax1);

plt.title('Actual vs Fitted Values for Price')
plt.legend()
plt.show()
```
